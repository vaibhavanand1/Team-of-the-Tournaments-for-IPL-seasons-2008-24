```python
import os
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
# !pip3 install googlesearch googlesearch-python
from googlesearch import search

import warnings
warnings.filterwarnings("ignore")
```

```python
file_path = os.path.join(r"C:\Users\vaibh\Documents\JobIntern\Projects\IPLTeamOfTournaments\Code\players\allstar.xlsx")
df = pd.read_excel(file_path)
seasons = ['2007/08', '2009', '2009/10', '2011', '2012', '2013', '2014', '2015', '2016',
 '2017', '2018', '2019', '2020/21', '2021', '2022', '2023', '2024']
players = []
for season in seasons:
    players = players + df[season].to_list()
del df
print(len(players))
```

```python
dt = {}
for player in players:
    dt[player] = False
    
path_photos = os.path.join(r"C:\Users\vaibh\Documents\JobIntern\Projects\IPLTeamOfTournaments\Code\playerphotos")
photos = os.listdir(path_photos)
for photo in photos:
    name = photo.split('.')[0].replace("_", " ")
    dt[name] = True
```

```python
for key, val in dt.items():
    if not val:
        print(f"Player photo to download: {key}")
```

```python
#### CRICBUZZ

folder_path = os.path.join(r"C:\Users\vaibh\Documents\JobIntern\Projects\IPLTeamOfTournaments\Code\playerphotos")

for player in players:
    if dt[player]: continue
    # if player == 'SM Curran': continue
    print(f'This is {player}')
    query = f'{player} cricbuzz profile'
    
    search_result = search(query, sleep_interval=10, num_results=1)
    first_url = next(search_result)
    
    # Introduce a delay before making a request
    time.sleep(10)
    
    # Fetch the profile page
    response = requests.get(first_url)
    
    # Check if the request was successful
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract the image URL
        img_tag = soup.find('div', class_='cb-col cb-col-20 cb-col-rt').find('img')
        if img_tag:
            image_url = img_tag['src']
            
            # Define the file path
            file_path = os.path.join(folder_path, f"{player.replace(' ', '_')}.jpg")
            
            # Introduce another delay before fetching the image
            time.sleep(5)
            
            # Fetch the image
            img_response = requests.get(image_url)
            
            # Check if the image request was successful
            if img_response.status_code == 200:
                with open(file_path, 'wb') as file:
                    file.write(img_response.content)
                print(f'Image downloaded successfully and saved as {file_path}!')
                dt[player] = True
            else:
                print(f'Failed to retrieve the image. Status code: {img_response.status_code}')
        else:
            print(f'No image found for {player}')
    else:
        print(f'Failed to retrieve the profile page. Status code: {response.status_code}')
    
    # Sleep after every request to avoid hitting rate limits
    time.sleep(15)
```
